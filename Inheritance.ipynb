# **Inheritence:-**

Inheritance allows us to define a class that inherits all the methods and properties from another class.
Parent class is the class being inherited from, also called base class. Child class is the class that inherits from another class, also called derived class. Any class can be a parent class, so the syntax is the same as creating any other class.


## Types of Inheritance--

*   Single 

*   Multilevel

*   Multiple
*   Hierarchical


*   Multipath


*   Hybrid







## Single Inheritance-

When a child class inherits only a single parent class.

Single.JPG

```python
#Example
class Phone:
  def __init__(self):
    pass
  def dial(self,num):
    print("Calling..",num)

class SmartPhone(Phone):
  def __init__(self):
    pass
    Phone.__init__(self)
  def msg(self):
    print("Sending SMS")
obj=SmartPhone()
obj.dial(7985919411)
```

# **Multilevel-**

In the Multilevel inheritance, a derived class will inherit a base class and as well as the derived class also act as the base class to other class. For example, three classes called A, B, and C, as shown in the below image, where class C is derived from class B and class B, is derived from class A.


Multilevel.JPG



```python
#Example
class Phone:
  def __init__(self):
    pass
  def call(self,num):
    print("Calling..",num)
  def msg(self,num):
    print("Sending msg",num)


class SmartPhone(Phone):
  def __init__(self):
    pass
    Phone.__init__(self)
  def camera(self):
      print("Opening Camera")
  def music(self):
      print("Playing music.....")
  def internet(self):
      print("browseing...")

class Android(SmartPhone):
  def __init__(self):
    pass
    SmartPhone.__init__(self)
  def chrome(self):
    print("Opening Chrome...")
  def pubG(self):
    print("Starting pubg game...")
  
obj=Phone()
obj1=SmartPhone()
 # obj2=Android()
obj2=Android()
obj.call(7985919411)
obj2.call(7985919411)
obj2.pubG()
```

## Multiple Inheritance-


Multiple inheritance means that a subclass can inherit from two or more superclasses. C++ allows multiple inheritance, but Java allows only single inheritance, that is, a subclass can inherit only one superclass.

Multiple.JPG

```python
class Phone:
  def __init__(self):
    pass
  def call(self,num):
    print("Calling..",num)
  def msg(self,num):
    print("Sending msg",num)


class SmartPhone:
  def __init__(self):
    pass
  def camera(self):
      print("Opening Camera")
  def music(self):
      print("Playing music.....")
  def internet(self):
      print("browseing...")

class Android:
  def __init__(self):
    pass
  def chrome(self):
    print("Opening Chrome...")
  def pubG(self):
    print("Starting pubg game...")
  
class Tablet(Phone,SmartPhone,Android):
  def __init__(self):
    pass
    Phone.__init__(self)
    SmartPhone.__init__(self)
    Android.__init__(self)
  def msWord(self):
    print("Starting MS Word...")
  def msExcel(self):
    print("Starting MS Excel...")
obj=Tablet()
obj.call(7985919411)
obj.msg(9984498297)
```

## Hierarchical Inheritance-

 When more than one derived classes are created from a single base this type of inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and two child (derived) classes.


 Hierarchical.JPG

```python
class Phone:
  def __init__(self):
    pass
  def call(self,num):
    print("Calling..",num)
  def msg(self,num):
    print("Sending msg",num)


class SmartPhone(Phone):
  def __init__(self):
    pass
    Phone.__init__(self)
  def camera(self):
      print("Opening Camera")
  def music(self):
      print("Playing music.....")
  def internet(self):
      print("browseing...")

class Android(Phone):
  def __init__(self):
    pass
    Phone.__init__(self)
  def chrome(self):
    print("Opening Chrome...")
  def pubG(self):
    print("Starting pubg game...")
  
class Tablet(SmartPhone):
  def __init__(self):
    pass
    SmartPhone.__init__(self)
  def msWord(self):
    print("Starting MS Word...")
  def msExcel(self):
    print("Starting MS Excel...")
obj=Phone()
obj1=SmartPhone()
obj2=Android()
obj3=Tablet()
obj.call(7985919411)
obj3.msg(9984498297)
```

## Multipath Inheritance-

When a class is derived from two or more classes which are derived from the same base class then such type of inheritance is called multipath inheritance


Multipath.JPG


```python
class Computer:
  def __init__(self):
    pass
  def on(self):
    print("Your system is opening...")
  def off(self):
    print("Your system is starting shuting down..")
  def cmd(self):
    print("Starting CMD Prompt.")

class UpgradedLaptop:
  def __init__(self):
    pass
  def windows10(self):
    print("Starting Wondows 10")
  def wifi(self):
    print("Ready to add network...")
  def bluetooth(self):
    print("Ready To Pair")
  def fingerprint(self):
    print("ready to access")

class Desktop(Computer,UpgradedLaptop):
  def __init__(self):
    pass
    Computer.__init__(self)
    UpgradedLaptop.__init__(self)
  def windows98(self):
    print("Starting windows 98")
  def notepad(self):
    print("Starting Notepad")
  def myComputer(self):
    print("Opening My computer")

class UpgradedDesktop(Computer,UpgradedLaptop):
  def __init__(self):
    pass
    Computer.__init__(self)
    UpgradedLaptop.__init__(self)
  def windows2000(self):
    print("Starting windows 2000")
  def paint(self):
    print("Starting paint")
  def controlPanel(self):
    print("Opening control Panel")

class Laptop(Computer,UpgradedLaptop):
  def __init__(self):
    pass
    Computer.__init__(self)
    UpgradedLaptop.__init__(self)
  def msWord(self):
    print("Starting MS word")
  def msExcel(self):
    print("Starting MS Excel")
  def msppt(self):
    print("Starting Powerpoint")
  

  
obj=Desktop()
obj.notepad()
obj1=Laptop()
obj1.windows10()
```
