# **# What are Conditional Statements in Python?**

Conditional Statement in Python perform different computations or actions depending on whether a specific Boolean constraint evaluates to true or false. Conditional statements are handled by IF statements in Python. There are four types of conditional statements.

*   IF condition

*   IF-ELSE
*   ELIF Ladder


*   Switch



# **IF condtion**-  
 if Statement is used for decision-making operations. It contains a body of code which runs only when the condition given in the if statement is true.

```python
# Use of if
a=8
if a%2==0:
     print("action complete")
```

# **IF-Else Statement-**
If-else Statement is used for decision-making operations. It contains a body of code which runs only when the condition given in the if statement is true. If the condition is false, then the optional else statement runs which contains some code for the else condition.

Flowchart-of-If-Else-Statement.png

```python
# Program to check eligibility of voting
a=int(input("Enter your age"))
if a>=18:
       print("You are eligible")
else:
        print("You are illigible")
```

# **Elif Ladder-**
The elif statement allows you to check multiple expressions for TRUE and execute a block of code as soon as one of the conditions evaluates to TRUE. Similar to the else, the elif statement is optional.

Python_if_elif_else_statement.jpg

```python
# Program to calculate division by your percentage
per=float(input("Enter your percentage"))
if per>=60 and per<=100:
    print("you are passed with 1st division")
elif per<60 and per>=50:
    print("You are passed with 2nd division")
elif per<50 and per>=35:
    print("You are passed with 3rd division")
else:
    print(" Yor are failed")
```
